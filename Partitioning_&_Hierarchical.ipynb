```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
# Load Iris dataset

iris = load_iris()
data_iris = iris.data
```

```python
# Standardize the data for K-Means
scaler = StandardScaler()
data_iris_standardized = scaler.fit_transform(data_iris)
```

```python
# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(data_iris_standardized)
```

```python
# Hierarchical clustering
linked = linkage(data_iris_standardized, 'ward')
```

```python
# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(data_iris_standardized[:, 0], data_iris_standardized[:, 1], c=kmeans_labels, cmap='viridis')
plt.title("K-Means Clustering - Iris Dataset")

plt.subplot(1, 2, 2)
dendrogram(linked)
plt.title("Hierarchical Clustering - Iris Dataset")

plt.show()
```

```python
# Evaluate the performance with silhouette score for K-Means
silhouette_kmeans = silhouette_score(data_iris_standardized, kmeans_labels)
print(f"Silhouette Score (K-Means): {silhouette_kmeans}")
```

```python
titanic_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_df = pd.read_csv(titanic_url)
data_titanic = titanic_df[['Age', 'Fare']].dropna().values
data_titanic_standardized=data_titanic
```

```python
# K-Means clustering
kmeans_titanic = KMeans(n_clusters=2, random_state=42)
kmeans_labels_titanic = kmeans_titanic.fit_predict(data_titanic_standardized)
```

```python
# Hierarchical clustering
linked_titanic = linkage(data_titanic_standardized, 'ward')
```

```python
# Plot the results for Titanic dataset
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(data_titanic_standardized[:, 0], data_titanic_standardized[:, 1], c=kmeans_labels_titanic, cmap='viridis')
plt.title("K-Means Clustering - Titanic Dataset")

plt.subplot(1, 2, 2)
dendrogram(linked_titanic)
plt.title("Hierarchical Clustering - Titanic Dataset")

plt.show()
```

```python
# Evaluate the performance with silhouette score
silhouette_kmeans_titanic = silhouette_score(data_titanic_standardized, kmeans_labels_titanic)
print(f"Silhouette Score (K-Means) - Titanic Dataset: {silhouette_kmeans_titanic}")
```
